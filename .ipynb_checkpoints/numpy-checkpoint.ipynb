```python
expenses = np.array([
    [200,340,230,150,456],
    [100,240,433,234,233],
    [234,123,433,444,345]
])
```

```python
print('total expense: ', np.sum(expenses))
print('category wise sum: ',np.sum(expenses,axis=1))
print('day wise total: ',np.sum(expenses ,axis = 0))
print('highest expense day:',np.max(np.sum(expenses ,axis=0)))
```

```python
import numpy as np
marks = np.array([34,45,54,78,87,65,55])
```

```python
grades = np.where(marks>=75,"A",
                  np.where(marks>=55 ,"B",
                  np.where(marks>=40,"C","fail")))
print(grades)
                  
```

```python
marks
```

```python
print(marks>60)
```

```python
print(marks[marks>60])
```

```python
print(marks[(marks>60) &(marks>80)])
```

```python
print(marks[(marks>60) | (marks>80)])
```
